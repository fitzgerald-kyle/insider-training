## sklearn continuous outcome

```python
delta_train = datetime.datetime.strptime('2021-06-19', '%Y-%m-%d') - datetime.datetime.strptime('2021-06-01', '%Y-%m-%d')

train_days = []
for i in range(delta_train.days + 1):
    train_days.append(datetime.datetime.strptime('2021-06-01', '%Y-%m-%d') + datetime.timedelta(days=i))

delta_test = datetime.datetime.strptime('2021-06-30', '%Y-%m-%d') - datetime.datetime.strptime('2021-06-20', '%Y-%m-%d')

test_days = []
for i in range(delta_test.days + 1):
    test_days.append(datetime.datetime.strptime('2021-06-20', '%Y-%m-%d') + datetime.timedelta(days=i))

insiderDat.TradeDate = [datetime.datetime.strptime(d, '%Y-%m-%d %H:%M:%S') for d in insiderDat.TradeDate]
insiderDat_purchases = insiderDat[insiderDat.TradeType == 'P - Purchase'] #only interested in purchases
insiderDat_purchases.drop(columns=['TradeType','CompanyName','Unnamed: 0', 'X', 'FilingDate', '1d', '1w', '1m', '6m', 'ref', 
                                   'changeCat'], 
                          inplace=True)
```

```python
# Making training and testing sets
filtered_data = insiderDat_purchases.drop(columns=['Ticker', '10%', '50%', '100%', 'InsiderName'])
dummies_data = pd.get_dummies(filtered_data, columns=["Title"], prefix=["Title_"])

training_data_XY = dummies_data[dummies_data['TradeDate'].isin(train_days)]
training_data_XY.dropna(inplace=True)
training_data_X = training_data_XY.drop(columns=['perChange', 'TradeDate'])
training_Y = training_data_XY['perChange']

assert np.any(np.isnan(training_data_X)) == False
assert np.all(np.isfinite(training_data_X)) == True
assert np.any(np.isnan(training_Y)) == False
assert np.all(np.isfinite(training_Y)) == True

testing_data_XY = dummies_data[dummies_data['TradeDate'].isin(test_days)]
testing_data_XY.dropna(inplace=True)
testing_data_X = testing_data_XY.drop(columns=['perChange', 'TradeDate'])
testing_Y = testing_data_XY['perChange']

assert np.any(np.isnan(testing_data_X)) == False
assert np.all(np.isfinite(testing_data_X)) == True
assert np.any(np.isnan(testing_Y)) == False
assert np.all(np.isfinite(testing_Y)) == True
```

```python
training_data_X.head()
```

```python
skr_parameters = [{
'max_depth': [3, 5, 10, None], 
'min_samples_split': [2, 5, 10],
'min_samples_leaf': [1, 5, 10]}]

dtr = GridSearchCV(
        tree.DecisionTreeRegressor(), skr_parameters
    )
dtr.fit(training_data_X, training_Y)
print(dtr.best_params_)
```

```python
skr = tree.DecisionTreeRegressor(max_depth= 5, min_samples_leaf= 10, min_samples_split= 2)
skr = skr.fit(training_data_X, training_Y)
```

```python
preds = skr.predict(testing_data_X)
metrics.mean_squared_error(testing_Y, preds)
```

```python
# export the decision tree to a tree.dot file
# for visualizing the plot easily anywhere
export_graphviz(skr, out_file ='tree.dot',
               feature_names = training_data_X.columns) 
```

```python
skr_importances = pd.Series(skr.feature_importances_, index=training_data_X.columns)

fig, ax = plt.subplots()
skr_importances.plot.bar()
ax.set_title("Feature importances using Regression tree")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(skr, feature_names=training_data_X.columns, rounded=True, filled=True)
fig.savefig("regression_tree.png")
```

## Strategy

```python
full_test_dat_preds = testing_data_XY.copy()
full_test_dat_preds['preds'] = preds
full_test_dat_preds.head()
```

```python
def build_flag(r):
    if r >= 0.20 and r < 0.50:
        return 'between 20 and 50%'
    elif r >= 0.50 and r < 1:
        return 'between 50% and 100%'
    elif r >= 1:
        return 'at least 100%'
    else:
        return 'not enough'

full_test_dat_preds['flag'] = [build_flag(r) for r in full_test_dat_preds.preds]
full_test_dat_preds['flag'].value_counts()
```

```python
TickerDat = pd.read_pickle('data/tickerDat_fmt.pkl')
```

### Strategy: at least 20%

Constraints: buy stock one day after TradeDate to allow for filing time (usually same day)

```python
purs_to_mimick_20 = full_test_dat_preds.copy()
purs_to_mimick_20 = purs_to_mimick_20[purs_to_mimick_20['flag'] != 'not enough']
purs_to_mimick_20['Pur_day'] = [t + datetime.timedelta(days=1) for t in purs_to_mimick_20.TradeDate]
purs_to_mimick_20 = purs_to_mimick_20.join(insiderDat_purchases['Ticker'])
purs_to_mimick_20['weekday'] = [d.weekday() for d in purs_to_mimick_20.Pur_day]
purs_to_mimick_20['weekday'].value_counts() # can't be 5 (Sat) or 6 (Sund)
```

```python
def find_value(i):
    try:
        t = TickerDat[purs_to_mimick_20['Ticker'][i]][TickerDat[purs_to_mimick_20['Ticker'][i]].index == purs_to_mimick_20['Pur_day'][i]]['Open'][0]
        return t
    except:
        print('Didn t work for these indices: {}'.format(i))
        
purs_to_mimick_20['stockPrice_pur'] = [find_value(i) for i in purs_to_mimick_20.index]
```

```python
purs_to_mimick_20['stockPrice_pur'][1001] = TickerDat['FTCV'][TickerDat['FTCV'].index == '2021-07-01']['Open'][0]
purs_to_mimick_20['stockPrice_pur'][1036] = TickerDat['WWR'][TickerDat['WWR'].index == '2021-07-01']['Open'][0]
purs_to_mimick_20['stockPrice_pur'][1041] = TickerDat['VINC'][TickerDat['VINC'].index == '2021-07-01']['Open'][0]
purs_to_mimick_20['stockPrice_pur'][1043] = TickerDat['VINC'][TickerDat['VINC'].index == '2021-07-01']['Open'][0]
```

```python
# building portfolio -- buying 5 of each stock (what's important is % change anyways)
purs_to_mimick_20['pur_value'] = [p*5 for p in purs_to_mimick_20.stockPrice_pur]
sum(purs_to_mimick_20['pur_value']) #how much did we spend?
```

```python
# selling when value hits at least +20% over next month
def sale_price(i): #index of purs df
    price_to_beat = purs_to_mimick_20['stockPrice_pur'][i]*1.20 # increase of 20%
    tDat = TickerDat[purs_to_mimick_20['Ticker'][i]]
    tDat_flt = tDat[tDat.index > purs_to_mimick_20['Pur_day'][i]]
    tDat_flt_p = tDat_flt[tDat_flt['Open'] > price_to_beat]
    try:
        return [tDat_flt_p['Open'].iloc[0], tDat_flt_p.index[0]]
    except:
        print('These tickers don t get over 20% change: {}'.format(purs_to_mimick_20['Ticker'][i]))
        return [tDat['Open'].iloc[-1], tDat.index[-1]] # selling them at the end of our 3 months timeline

sales_dat = pd.DataFrame([sale_price(i) for i in purs_to_mimick_20.index], columns=['sale_price', 'sale_date'])
purs_to_mimick_20['sale_price'] = np.array(sales_dat.sale_price)
purs_to_mimick_20['sale_date'] = np.array(sales_dat.sale_date)
```

```python
purs_to_mimick_20['sale_value'] = purs_to_mimick_20.sale_price*5 #number of stocks we bought
total_per_change = (sum(purs_to_mimick_20.pur_value) - sum(purs_to_mimick_20.sale_value))/sum(purs_to_mimick_20.pur_value)
total_per_change*100
```
